**Color Maps in OpenCV**
- We are going to see about color maps and some programs using colormaps.

**What is color map ?**
- Color maps is a feature in OpenCV that helps us to visualise images in a predefined/custom color scheme in images, especially grayscale images.

**Why need color maps?**
- Suppose, There is a image of Map of India or The Globe in grayscale where each color represents different temperature. Do you think, is it easy for our eyes to note the variations in the temperature in grayscale?

Difficult to answer (or) don't you understand what am I saying ? 
Let's see an image of India.

```python
#importing libraries 
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
image_1 = cv2.imread(r'E:\Learning_skills\Computer_vision\color_maps\india_map.jpg')
plt.imshow(image_1)
```

The above image is a grayscale image of India map with temperature in different intesities at different areas.
Do you think from the image, you can identify the variations easily. I think No!

Suppose we have a magic tool that makes the grayscale in a color scheme, It can make our eyes easy to identify the temperature variations.
That **Magic tool** is **Color maps**.

Let's see how will image will look like 

```python
im_color = cv2.applyColorMap(image_1, cv2.COLORMAP_JET)
im_color = cv2.cvtColor(im_color, cv2.COLOR_BGR2RGB)
plt.imshow(im_color)
```

**The image looks like some how good to our eyes than the grayscale image, right?**

It doesn't only useful in visualisation part only. It also helps us to find many pattern, features that are very important in many fields like healthcare, Thermal Imaging and so on.

**List of colormaps available:**

**Some other different color maps implemetation:**

```python
#example 1
image_2 = cv2.imread(r'E:\Learning_skills\Computer_vision\color_maps\image_3.png')
img_map_2 = cv2.applyColorMap(image_2, cv2.COLORMAP_JET)

plt.subplot(1,2,1)
plt.imshow(image_2)
plt.title('Orginal Image')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(img_map_2)
plt.title('Color map Image')
plt.axis('off')

plt.show()
```
