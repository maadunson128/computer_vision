First, we are going to find the range of the trunk from the image.

```python
bgr_values = []
hsv_values = []
```

```python
#function for getting the pixel values from the mouse left double click
def mouse_value(event, x, y, flags, image):
    if event == cv2.EVENT_LBUTTONDBLCLK:
        global bgr_values, hsv_values

        bgr_values.append(image[0][y, x])
        hsv_values.append(image[1][y, x])
        
        
        
```

```python
image_1_bgr = cv2.imread(r'E:\Learning_skills\Computer_vision\data\greenery\image_1.jpg', 1)
image_1_hsv = cv2.cvtColor(image_1_bgr, cv2.COLOR_BGR2HSV)
cv2.imshow('Image 1',image_1_bgr)
#cv2.setMouseCallback("Original",mouse_display)

cv2.setMouseCallback('Image 1', mouse_value, [image_1_bgr, image_1_hsv])
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#For image_1, we are going to find the range of bgr and hsv values
def range(bgr_values, hsv_values):
    bgr_array = np.array(bgr_values)
    hsv_array = np.array(hsv_values)

    #min and max values
    bgr_min = np.min(bgr_array, axis=0)
    bgr_max = np.max(bgr_array, axis = 0)

    hsv_min = np.min(hsv_array, axis=0)
    hsv_max = np.max(hsv_array, axis=0)

    print(f"BGR min value: {bgr_min} ")
    print(f"BGR max value: {bgr_max} ")

    print(f"HSV min value: {hsv_min} ")
    print(f"HSV max value: {hsv_max} ")
```

```python
#for image 2, finding the range
bgr_values = []
hsv_values = []

image_2_bgr = cv2.imread(r'E:\Learning_skills\Computer_vision\data\greenery\image_2.jpg', 1)
image_2_hsv = cv2.cvtColor(image_2_bgr, cv2.COLOR_BGR2HSV)
cv2.imshow('Image 2',image_2_bgr)
#cv2.setMouseCallback("Original",mouse_display)

cv2.setMouseCallback('Image 2', mouse_value, [image_2_bgr, image_2_hsv])
cv2.waitKey(0)
cv2.destroyAllWindows()
```
