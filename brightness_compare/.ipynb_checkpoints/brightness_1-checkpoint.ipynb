**1. BGR color space**

**Image 1**

```python
#loading image 1 using cv
image_1_bgr = cv2.imread('E:/Learning_skills/Computer_vision/data/brightness/image_6.jpg', 1)

#Printing the array representation of image 1 in BGR space
print(image_1_bgr)

#finding the mean of B, G, R colors
image_1_bgr_mean = np.mean(image_1_bgr, axis = 1).mean(axis = 0)

#printing the mean values
print(f"\n\nBlue color mean: {image_1_bgr_mean[0]} \nGreen color mean: {image_1_bgr_mean[1]} \nRed color mean: {image_1_bgr_mean[2]}")
```

Image 1 grayscale value

```python
0.299*image_1_bgr_mean[2]+0.587*image_1_bgr_mean[1]+0.114*image_1_bgr_mean[0]
```

**Image 2**

```python
#loading image 1 using cv
image_2_bgr = cv2.imread('E:/Learning_skills/Computer_vision/data/brightness/image_7.jpg', 1)

#Printing the array representation of image 1 in BGR space
print(image_2_bgr)

#finding the mean of B, G, R colors
image_2_bgr_mean = np.mean(image_2_bgr, axis = 1).mean(axis = 0)

#printing the mean values
print(f"\n\nBlue color mean: {image_2_bgr_mean[0]} \nGreen color mean: {image_2_bgr_mean[1]} \nRed color mean: {image_2_bgr_mean[2]}")
```

Image 2 grayscale value

```python
0.299*image_2_bgr_mean[2]+0.587*image_2_bgr_mean[1]+0.114*image_2_bgr_mean[0]
```

**2.HSV color space** 

**Image 1**

```python
#loading image 1 using cv
image_1_hsv = cv2.cvtColor(image_1_bgr, cv2.COLOR_BGR2HSV)

#Printing the array representation of image 1 in BGR space
print(image_1_hsv)

#finding the mean of B, G, R colors
image_1_hsv_mean = np.mean(image_1_hsv, axis = 1).mean(axis = 0)

#printing the mean values
print(f"\n\nHue mean: {image_1_hsv_mean[0]} \nSaturation mean: {image_1_hsv_mean[1]} \nLightness value mean: {image_1_hsv_mean[2]}")
```

**Image 2**

```python
#loading image 1 using cv
image_2_hsv = cv2.cvtColor(image_2_bgr, cv2.COLOR_BGR2HSV)

#Printing the array representation of image 1 in BGR space
print(image_2_hsv)

#finding the mean of B, G, R colors
image_2_hsv_mean = np.mean(image_2_hsv, axis = 1).mean(axis = 0)

#printing the mean values
print(f"\n\nHue mean: {image_2_hsv_mean[0]} \nSaturation mean: {image_2_hsv_mean[1]} \nLightness value mean: {image_2_hsv_mean[2]}")
```

**3. YUV color space**

**Image 1**

```python
#loading image 1 using cv
image_1_yuv = cv2.cvtColor(image_1_bgr, cv2.COLOR_BGR2YUV)

#Printing the array representation of image 1 in BGR space
print(image_1_yuv)

#finding the mean of B, G, R colors
image_1_yuv_mean = np.mean(image_1_yuv, axis = 1).mean(axis = 0)

#printing the mean values
print(f"\n\nLuma (Brightness) value mean: {image_1_yuv_mean[0]} \nChroma value (U) mean: {image_1_yuv_mean[1]} \nChroma (V) mean: {image_1_yuv_mean[2]}")
```

**Image 2**

```python
#loading image 1 using cv
image_2_yuv = cv2.cvtColor(image_2_bgr, cv2.COLOR_BGR2YUV)

#Printing the array representation of image 1 in BGR space
print(image_2_yuv)

#finding the mean of B, G, R colors
image_2_yuv_mean = np.mean(image_2_yuv, axis = 1).mean(axis = 0)

#printing the mean values
print(f"\n\nLuma (Brightness) value mean: {image_2_yuv_mean[0]} \nChroma value (U) mean: {image_2_yuv_mean[1]} \nChroma (V) mean: {image_2_yuv_mean[2]}")
```
